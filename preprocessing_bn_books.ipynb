```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import langdetect as ld
```

```python
df = pd.read_csv("books_modified_1.csv")
```

```python
df.head(5)
```

# Only Bangla Books

```python
df_bn = df.loc[df['language'] == 'bn']
df_bn.reset_index(drop=True, inplace=True)
```

```python
df_bn.head(5)
```

```python
def detect_language(row):
    try:
        lang = ld.detect(str(row))
    except:
        return 'unknown'
    return lang
```

```python
df_bn['language_author'] = df_bn['author'].apply(detect_language)
df_bn['language_categories'] = df_bn['categories'].apply(detect_language)
```
